## LOAD *Iris* data

```python
# load dataset
dataset = load_iris()
dataset.keys()
# dataset to df
data = pd.DataFrame(dataset.data, columns = dataset.feature_names)
data['class'] = dataset.target
dclass = dict()
for i, ic in enumerate(dataset.target_names):
    dclass[i] = ic
data['class'] = data['class'].map(dclass)
```

# HISTOGRAMS

```python
# collect num columns
columns_num = data.select_dtypes(include=['float64']).columns.tolist() 
```
